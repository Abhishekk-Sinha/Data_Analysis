```python
import plotly.graph_objects as go
```

```python
# !pip install plotly
#add_trace is method used to add one or more traces to a plot. Traces represent indiidual sets of data and other aspects of plot
#that can be visualised on the same plot

fig = go.Figure()
fig.add_trace(go.Scatter(x = [1, 2,3,4,5], y = [3,4,5,8,9], mode = "markers" ))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = [1,2,3,4,5], y = [3,4,5,7,1], mode = "lines"))
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill, y = tips.tip, mode = "markers"))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill, y = tips.tip, mode = "lines"))
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x = [1,2,3,4,5], y = [3,5,7,2,9]))
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=tips['total_bill']))  #to visualise the continuous data
```

```python
tips
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill, y = tips.tip, mode = "markers", marker_size = 12*tips['size']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = tips.total_bill, y = tips.tip, mode = "markers", marker_size = 12*tips['size'], z = tips['size']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = [1,2,3,4,1], y = [4,5,6,7,8], mode = 'lines', z = [8,3,2,6,8]))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = [1,2,3,4,5], y = [4,2,7,1,6], z = [9,7,3,1,4], mode = 'markers'))
```
