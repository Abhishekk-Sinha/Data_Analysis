- Q. The difference between Deep copy and shallow copy.

- Deep copy doesn't reflect changes made to the new/copied object in the original object.
- Shallow Copy stores the copy of the original object and points the references to the objects.
- Deep copy stores the copy of the original object and recursively copies the objects as well.
- A shallow copy is faster.

- Q.15. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:

  - a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

  - b) Label the x-axis as 'Date'.

  - c) Set the title of the plot as 'Temperature and Humidity Over Time'.


```python
# Creating a time-series dataset
data = {
    'Date': pd.date_range(start='2024-01-01', periods=100),
    'Temperature': [25 + i * 0.5 for i in range(100)],  # Generating temperature data
    'Humidity': [50 + i * 0.3 for i in range(100)]      # Generating humidity data
}

# Creating DataFrame
df = pd.DataFrame(data)

# Plotting
fig, ax1 = plt.subplots()

# Plotting Temperature
color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (Â°C)', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Creating another y-axis for Humidity
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity (%)', color=color)
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Title
plt.title('Temperature and Humidity Over Time')

# Rotating x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Displaying the plot
plt.show()
```

Q.17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
np.random.seed(42)
x = np.random.normal(size=100)
y = np.random.normal(size=100)

# Determine quadrant
quadrant = np.zeros_like(x, dtype=int)
quadrant[(x > 0) & (y > 0)] = 1
quadrant[(x < 0) & (y >= 0)] = 2
quadrant[(x < 0) & (y < 0)] = 3
quadrant[(x >= 0) & (y < 0)] = 4

# Create a dataframe
data = {'x': x, 'y': y, 'quadrant': quadrant}
df = pd.DataFrame(data)

# Set style
sns.set(style="whitegrid")

#plotting
plt.figure(figsize=(8, 6))
sns.scatterplot(x='x', y='y', hue='quadrant', data=df, palette='viridis')

# Labeling axes
plt.xlabel('X')
plt.ylabel('Y')

# Set title
plt.title('Quadrant-wise Scatter Plot')

# Display legend
plt.legend(title='Quadrant', loc='upper right')

plt.show()
```

18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Generate data for sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

#Output to notebook

output_notebook

# Create a figure
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='y')


# Add a line renderer
p.line(x, y, line_width=2)

#add grid lines
p.xgrid.visible = True
p.ygrid.visible = True

show(p)
```

- 19.  Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'.

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6
import numpy as np
import pandas as pd

# Output to notebook
output_notebook()

# Generate random categorical data
np.random.seed(0)
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(categories))

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a figure
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values', width=800, height=400)

# Create bar chart
p.vbar(x='categories', top='values', width=0.9, source=source,
       legend_field='categories', line_color='white', fill_color=factor_cmap('categories', palette=Spectral6, factors=categories))

# Add hover tool
hover = HoverTool()
hover.tooltips = [
    ("Category", "@categories"),
    ("Value", "@values")
]
p.add_tools(hover)

# Customize legend
p.legend.orientation = 'horizontal'
p.legend.location = 'top_center'

# Show plot
show(p)
```

Q.20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'8

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(0)
x = np.arange(0, 100)
y = np.random.randn(100).cumsum()

# Create line plot
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines', name='Line Plot'))

# Label the axes
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show plot
fig.show()
```

Q.21. Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'.

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(0)
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(categories))

# Create pie chart
fig = go.Figure(data=[go.Pie(labels=categories, values=values, hole=0.3,
                             hoverinfo='label+percent', textinfo='label+percent',
                             textfont_size=14)])

# Set title
fig.update_layout(title_text='Interactive Pie Chart')

# Show plot
fig.show()
```
